```python
import os
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
from sklearn.naive_bayes import BernoulliNB,MultinomialNB
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_dataset():
    directory = 'dataset'
    files = os.listdir(directory)
    classes = []
    for file in files:
        classes.append(file.rstrip('.tsv'))
    
    dfs = []
    for file in files:
        file_path = 'dataset/'+file
        df = pd.read_csv(file_path, sep='\t')
        dfs.append(df)
    df = pd.concat(dfs, ignore_index=True)
    df = df.astype(str)
    df = df.fillna("")
    return df
    
```

```python

def plotting(macroBNB, macroMNB, microBNB, microMNB, accBNB, accMNB):
    plt.figure(figsize=(14, 8))

    plt.subplot(2, 2, 1)
    plt.plot(macroBNB,color='r', marker='o', label='macroBNB')
    plt.plot(macroMNB,color='b', marker='o', label='macroMNB')

    plt.title('Macro Metrics')
    plt.xlabel('Fold')
    plt.ylabel('Score')
    plt.legend()
    plt.axhline(y=macroBNB.mean(), color='r', linestyle='--', label='macroBNB mean')
    plt.axhline(y=macroMNB.mean(), color='b', linestyle='--', label='macroMNB mean')

    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(microBNB,color='r', marker='o', label='microBNB')
    plt.plot(microMNB,color='b', marker='o', label='microMNB')

    plt.title('Micro Metrics')
    plt.xlabel('Fold')
    plt.ylabel('Score')
    plt.legend()
    plt.axhline(y=microBNB.mean(), color='r', linestyle='--', label='microBNB mean')
    plt.axhline(y=microMNB.mean(), color='b', linestyle='--', label='microMNB mean')

    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot(accBNB, color='r',marker='o', label='accBNB')
    plt.plot(accMNB,color='b', marker='o', label='accMNB')

    plt.title('Accuracy Metrics')
    plt.xlabel('Fold')
    plt.ylabel('Score')
    plt.legend()
    plt.axhline(y=accBNB.mean(), color='r', linestyle='--', label='accBNB mean')
    plt.axhline(y=accMNB.mean(), color='b', linestyle='--', label='accMNB mean')

    plt.legend()

    plt.tight_layout()
    plt.show()
```

# Orignal data preprocessing

```python
nltk.download('stopwords')
nltk.download('punkt')

df = get_dataset()
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))


def org_preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text) 
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [ps.stem(word) for word in tokens]
    return text
    
df['Release Year'] = df['Release Year'].astype(str)
df['Content'] = df.drop(columns=['Genre']).apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
df['Content'] = df['Content'].apply(org_preprocess_text)
```

```python
X = df['Content']
y = df['Genre']
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
pipelineBNB = Pipeline(steps=[
    ('preprocessor', CountVectorizer()),
    ('classifier', BernoulliNB())
])

pipelineMNB = Pipeline(steps=[
    ('preprocessor', CountVectorizer()),
    ('classifier', MultinomialNB())
])



macroBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='f1_macro')
macroMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='f1_macro')

microBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='f1_micro')
microMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='f1_micro')

accBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='accuracy')
accMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='accuracy')


plotting(macroBNB,macroMNB,microBNB,microMNB,accBNB,accMNB)
```

First, in this section, we experimented with the performance of MNB and BNB on the genre classification task under the original data preprocessing method. In this preprocessing method, we removed all symbols from the text, retaining only alphabetic characters and whitespace characters. Additionally, we removed words such as "is" and "this" that do not provide more information (stop words) for the CountVectorizer and may introduce potential noise. We then applied the StratifiedKFold cross-validation method. Since our data has an equal number of instances for each category, this cross-validation method can maintain an equal proportion of each category in the training set after shuffling the dataset, thus avoiding the issue of training set imbalance due to resampling.

The above figure shows the results of ten-fold cross-validation for the two models using the original data preprocessing method. The dashed line represents the average value of the respective metric for each model. We can see that the MNB model achieved higher scores than the BNB model for all three metrics, and the differences are not significant.

# Modified Data preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
df = get_dataset()

modify = r'[^\w\s\'-]'


def preprocess_name(text):
    text = text.lower()
    text = re.sub(modify, '', text)
    tokens = word_tokenize(text)
    return ' '.join(tokens)
    
def preprocess_text(text):
    text = text.lower()
    text = re.sub(modify, '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [ps.stem(word) for word in tokens]
    return ' '.join(tokens)

df['Release Year'] = df['Release Year'].astype(str)
for col in df.columns:
    if col != 'Genre':
        if col == 'Director' or col == 'Cast' or col == 'Title':
            df[col] = df[col].apply(preprocess_name)
        else:
            df[col] = df[col].apply(preprocess_text)
df['Content'] = df.drop(columns=['Genre']).apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
```

Here, we chose to retain the symbols ' and -, which better align with the grammatical requirements of the language and prevent hyphenated words from being split into two tokens. Since we are using CountVectorizer, a feature extraction method that does not include semantic analysis capabilities, retaining more punctuation is meaningless and may introduce potential noise (although it might improve the model's performance from an unusual perspective).

```python
X = df['Content']
y = df['Genre']
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
pipelineBNB = Pipeline(steps=[
    ('preprocessor', CountVectorizer()),
    ('classifier', BernoulliNB())
])

pipelineMNB = Pipeline(steps=[
    ('preprocessor', CountVectorizer()),
    ('classifier', MultinomialNB())
])

macroBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='f1_macro')
macroMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='f1_macro')

microBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='f1_micro')
microMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='f1_micro')

accBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='accuracy')
accMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='accuracy')


plotting(macroBNB,macroMNB,microBNB,microMNB,accBNB,accMNB)
```

The above figure shows the results of the two models on different metrics using the new data preprocessing method. Compared to the original preprocessing method, the new preprocessing method has brought a certain degree of improvement in the performance of both models. However, neither model is significantly better than the other.

Since our dataset is relatively balanced, the results for each metric do not differ significantly. For the genre classification task, we prefer a metric that focuses on overall accuracy. Therefore, we have chosen the Micro F1 score as the evaluation standard for the subsequent analysis.

# Apply top-K 

In this section, we primarily investigate the impact of changing the size of the vector by altering the max_features parameter of the CountVectorizer on the model's performance.

```python
meanBNB = []
meanMNB = []

vect = CountVectorizer()
features = vect.fit_transform(X)
for k in range(1,int(features.shape[1])+1,1000):
    pipelineBNB = Pipeline(steps=[
        ('preprocessor', CountVectorizer(max_features=k)),
        ('classifier', BernoulliNB())
    ])
    pipelineMNB = Pipeline(steps=[
        ('preprocessor', CountVectorizer(max_features=k)),
        ('classifier', MultinomialNB())
    ])
    microBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='f1_micro')
    microMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='f1_micro')

    meanBNB.append(microBNB.mean())
    meanMNB.append(microMNB.mean())

```

```python
k_values = list(range(1,int(features.shape[1])+1,1000))

plt.figure(figsize=(10, 6))
plt.plot(k_values, meanBNB, color='r', marker='o', label='BernoulliNB')
plt.plot(k_values, meanMNB,  color='b',marker='o', label='MultinomialNB')

plt.xlabel('k (max_features)')
plt.ylabel('Mean F1 Score (Micro)')
plt.title('Effect of k (max_features) on F1 Score (Micro)')
plt.legend()
plt.grid(True)
plt.show()
```

We adopted a brute-force search method, exploring the impact of max_features ranging from 1 to the complete set of features (increasing by 100 features each time due to limited computational resources) on the model's performance. The above figure shows the results of this brute-force search. When the max_features exceeds 5000, the improvement in model performance tends to stabilize and approaches 0.6. Therefore, we choose 10000 as the value for max_features.

```python
from sklearn.svm import SVC
k = 15000
pipelineBNB = Pipeline(steps=[
    ('preprocessor', CountVectorizer(max_features=k)),
    ('classifier', BernoulliNB())
])
pipelineMNB = Pipeline(steps=[
    ('preprocessor', CountVectorizer(max_features=k)),
    ('classifier', MultinomialNB())
])
pipelineSVM = Pipeline(steps=[
    ('preprocessor', CountVectorizer(max_features=k)),
    ('classifier', SVC())
])
microBNB = cross_val_score(pipelineBNB, X, y, cv=cv, scoring='f1_micro')
microMNB = cross_val_score(pipelineMNB, X, y, cv=cv, scoring='f1_micro')
microSVM = cross_val_score(pipelineSVM, X, y, cv=cv, scoring='f1_micro')
plt.figure(figsize=(10, 6))
plt.plot(microBNB, color='r', marker='o', label='microBNB')
plt.plot(microMNB, color='b', marker='o', label='microMNB')
plt.plot(microSVM, color='g', marker='o', label='microSVM')
plt.title('Micro Metrics')
plt.xlabel('Fold')
plt.ylabel('Score')
plt.legend()
plt.axhline(y=microBNB.mean(), color='r', linestyle='--', label='microBNB mean')
plt.axhline(y=microMNB.mean(), color='b', linestyle='--', label='microMNB mean')
plt.axhline(y=microSVM.mean(), color='g', linestyle='--', label='microSVM mean')
plt.legend()
```

The above figure shows the performance of the two models on the micro F1 metric when max_features is set to 15000. At the same time, we also evaluated and compared the performance of SVM with the same data preprocessing methods. Although this hyperparameter optimization did not significantly improve the overall performance of the model, compared to the previous cross-validation results, we can see that after hyperparameter optimization, the model's performance on each training set has become more stable. Additionally, the overall results of MNB are better than those of BNB and SVM. Therefore, we chose to use MNB with max_features set to 15000 to complete the genre classification task.

```python
# Function for getting training set (week 1-3) and test set (week 4)

def get_train_test():
    directory = 'dataset'
    files = os.listdir(directory)
    classes = []
    for file in files:
        classes.append(file.rstrip('.tsv'))
    train = []
    test = []
    for file in files:
        file_path = 'dataset/'+file
        df = pd.read_csv(file_path, sep='\t')
        train.append(df.iloc[:150])
        test.append(df.iloc[150:])
    train = pd.concat(train, ignore_index=True)
    test = pd.concat(test, ignore_index=True)

    return train, test
```

```python
# preprocess data
train, test = get_train_test()
def data_preprocess(df):
    df = df.astype(str)
    df = df.fillna("")
    for col in df.columns:
        if col != 'Genre':
            if col == 'Director' or col == 'Cast' or col == 'Title':
                df[col] = df[col].apply(preprocess_name)
            else:
                df[col] = df[col].apply(preprocess_text)
    df['Content'] = df.drop(columns=['Genre']).apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
    return df['Content']
train['Content'] = data_preprocess(train)
test['Content'] = data_preprocess(test)
```

```python
# read user data
user1 = pd.read_csv('user1.tsv', sep='\t')
user1.columns = ['Genre','Interests']
user2 = pd.read_csv('user2.tsv', sep='\t')
user2.columns = ['Genre','Interests']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# funciton for returning words with top 20 TF-IDF score
def get_top20(grouped):
    tfidf = TfidfVectorizer()
    tfidf.fit_transform(train['Content'])
    grouped_tfidf = tfidf.transform(grouped['Content']).toarray()
    
    vocab = tfidf.vocabulary_
    reverse_vocab = {v:k for k,v in vocab.items()}
    feature_names = tfidf.get_feature_names_out()
    df_tfidf = pd.DataFrame(grouped_tfidf, columns = feature_names)
    idx = grouped_tfidf.argsort(axis=1)
    tfidf_max20 = idx[:,-20:]
    df_tfidf['top20'] = [[reverse_vocab.get(item) for item in row] for row in tfidf_max20 ]
    
    grouped['top20'] = df_tfidf['top20']
    return grouped['top20']

# Function for matching user interests(keyword) with traning data
def user_match(user):
    tfidf = TfidfVectorizer()
    user['Content'] = user.iloc[:, [1]]
    user['Content'] = user['Content'].apply(preprocess_text)
    user_dict = dict(zip(user['Genre'], user['Content']))
    train["user_interests"] = train['preds'].map(user_dict)
    train['user_interests'] = train['user_interests'].fillna('')
    
    user_dict = dict(zip(user['Genre'], user['Content']))
    train['user_interests'] = train['preds'].map(user_dict)
    train['user_interests'] = train['user_interests'].fillna('')
    def check_match(row):
        content_set = set(row['Content'].split())
        interests_set = set(row['user_interests'].split())
        return int(len(content_set & interests_set) > 0)
    train['user_match'] = train.apply(check_match, axis=1)
    train['user_match'] = ((train['user_match'] > 0) & (train['preds'] == train['Genre'])).astype(int)
    
    result = train[train['user_match'] > 0]
    result = result[result['Genre'] == result['preds']]
    grouped = result.groupby('preds')['Content'].apply(' '.join).reset_index()
    grouped['top20'] = get_top20(grouped)
    group_dict = dict(zip(grouped['preds'], grouped['top20']))
    user['top20'] = user['Genre'].map(group_dict)
    user['profile'] = grouped['Content']
    return user
    
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)
```

# Recommendation

```python
from sklearn.neighbors import NearestNeighbors



def get_top_words(grouped,M):
    # =========================================
    # return top M word in user profile
    # =========================================
    tfidf = TfidfVectorizer()
    tfidf.fit_transform(train['Content'])
    grouped_tfidf = tfidf.transform(grouped['profile']).toarray()
    
    vocab = tfidf.vocabulary_
    reverse_vocab = {v:k for k,v in vocab.items()}
    feature_names = tfidf.get_feature_names_out()
    df_tfidf = pd.DataFrame(grouped_tfidf, columns = feature_names)
    idx = grouped_tfidf.argsort(axis=1)
    tfidf_max20 = idx[:,-M:]
    df_tfidf['top'] = [[reverse_vocab.get(item) for item in row] for row in tfidf_max20 ]
    grouped['top'] = df_tfidf['top']
    return grouped['top']
    
def filter_profile_words(user_profile, top_words):
    # =========================================
    # keep only words in Top M
    # =========================================
    profile_words = user_profile.split()
    filtered_words = [word for word in profile_words if word in top_words]
    return ' '.join(filtered_words)
    
def knn_recommender(user,method,N,M):
    
    # =========================================
    # define true labels in test data 
    # base on matching keywords
    # =========================================
    tfidf = TfidfVectorizer()
    user_dict = dict(zip(user['Genre'], user['Content']))
    test['user_interests'] = test['preds'].map(user_dict)
    test['user_interests'] = test['user_interests'].fillna('')
    def check_match(row):
        content_set = set(row['Content'].split())
        interests_set = set(row['user_interests'].split())
        return int(len(content_set & interests_set) > 0)
    test['user_match'] = test.apply(check_match, axis=1)
    test['user_match'] = ((test['user_match'] > 0) & (test['preds'] == test['Genre'])).astype(int)

    # =========================================
    # fit TfidfBectorizer on train data
    # =========================================
    tfidf = TfidfVectorizer()
    tfidf.fit_transform(train['Content'])

    # =========================================
    # vectorize test data and user profile
    # =========================================
    content_tfidf = tfidf.transform(test['Content'])
    user['top_words']=get_top_words(user,M)
    user['filtered_profile'] = user.apply(lambda row: filter_profile_words(row['profile'], row['top_words']), axis=1)
    interests_tfidf = tfidf.transform(user['filtered_profile'])

    
    # =========================================
    # fit KNN model on test data 
    # by using specific similarity method
    # =========================================
    knn = NearestNeighbors(n_neighbors=N, metric=method)
    knn.fit(content_tfidf)
    distances, indices = knn.kneighbors(interests_tfidf)

    
    # =========================================
    # find top N recommendation
    # =========================================
    result = []
    for i in range(0,len(distances)):
        for j in range(0,N):
            result.append((distances[i][j],indices[i][j]))
    result = sorted(result, key=lambda x: x[0])
    
    sorted_indices = [x[1] for x in result[:N]]
    result = test.copy()
    result = result.iloc[sorted_indices].copy()
    result['user_preds'] = 1
    return result

```

```python
def get_metrics(result):
    from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
    y_true = result['user_match']
    y_pred = result['user_preds']
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    return accuracy, precision
```

```python
N = 10
M = 2000
results = {
    'User': [],
    'Algorithm': [],
    'Accuracy': []
}
users = [user1, user2, user3]
algorithm = ['cosine', 'euclidean']
for i, user in enumerate(users, start=1):
    for metric in algorithm:
        result = knn_recommender(user, metric, N, M)
        accuracy,precision = get_metrics(result)
        results['User'].append(f'User{i}')
        results['Algorithm'].append(metric)
        results['Accuracy'].append(accuracy)
results_df = pd.DataFrame(results)
```
